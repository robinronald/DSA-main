```python
def move_zeroes(arr, n):
    index = 0  

    for i in range(n):
        if arr[i] != 0:
            arr[index] = arr[i]
            index += 1

    while index < n:
        arr[index] = 0
        index += 1

arr = [0, 1, 0, 3, 12]
n = len(arr)
move_zeroes(arr, n)
print(arr)
```

```python
def rotate_array(arr, n, k):
    k = k % n 
    for i in range(n - k // 2):
        arr[i], arr[n - k - 1 - i] = arr[n - k - 1 - i], arr[i]

    for i in range(k // 2):
        arr[n - k + i], arr[n - 1 - i] = arr[n - 1 - i], arr[n - k + i]

    for i in range(n // 2):
        arr[i], arr[n - 1 - i] = arr[n - 1 - i], arr[i]

arr = [1, 2, 3, 4, 5]
```

```python
def product_except_self(arr, n):
    left = [1] * n
    right = [1] * n
    result = [1] * n

    left[0] = 1
    for i in range(1, n):
        left[i] = left[i - 1] * arr[i - 1]

    right[n - 1] = 1
    for i in range(n - 2, -1, -1):
        right[i] = right[i + 1] * arr[i + 1]

    for i in range(n):
        result[i] = left[i] * right[i]

    return result


arr = [1, 2, 3, 4]
n = len(arr)
output = product_except_self(arr, n)
print(output) 
```

```python
def two_sum(arr, n, target):
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == target:
                return [i, j]
    return [-1, -1]


arr = [2, 7, 11, 15]
n = len(arr)
target = 9
print(two_sum(arr, n, target))
```

```python
def min_jumps(arr, n):
    if n <= 1:
        return 0
    if arr[0] == 0:
        return -1  

    jumps = 1
    max_reach = arr[0]
    steps = arr[0]

    for i in range(1, n):
        if i == n - 1:
            return jumps  
        
        max_reach = max_reach if max_reach > i + arr[i] else i + arr[i]
        steps -= 1

        if steps == 0:
            jumps += 1
            if i >= max_reach:
                return -1  
            steps = max_reach - i

    return -1  

arr = [6, 2, 4, 0, 5, 1, 1, 4, 2, 9]
n = len(arr)
print(min_jumps(arr, n))  
```

```python
def sort_colors(arr, n):
    low, mid, high = 0, 0, n - 1

    while mid <= high:
        if arr[mid] == 0: 
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:  
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

arr = [2, 0, 2, 1, 1, 0]
n = len(arr)
sort_colors(arr, n)
print(arr) 
```

```python
def trap_rain_water(arr, n):
    if n < 3:
        return 0

    left_max = [0] * n
    right_max = [0] * n
    water_trapped = 0

    left_max[0] = arr[0]
    for i in range(1, n):
        left_max[i] = left_max[i - 1] if left_max[i - 1] > arr[i] else arr[i]

    right_max[n - 1] = arr[n - 1]
    for i in range(n - 2, -1, -1):
        right_max[i] = right_max[i + 1] if right_max[i + 1] > arr[i] else arr[i]

    for i in range(n):
        min_height = left_max[i] if left_max[i] < right_max[i] else right_max[i]
        water_trapped += min_height - arr[i]

    return water_trapped

arr = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
n = len(arr)
print(trap_rain_water(arr, n))
```
