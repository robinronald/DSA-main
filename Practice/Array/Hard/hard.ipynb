```python
import numpy as np

employees = int(input("Enter the number of employees: "))
months = int(input("Enter the number of months: "))

sales = np.zeros((employees, months), dtype=int)

print("Enter sales data row-wise (each row represents an employee's monthly sales):")
for i in range(employees):
    for j in range(months):
        sales[i][j] = int(input(f"Sales for Employee {i+1}, Month {j+1}: "))

# Calculate total sales for each employee
employee_totals = np.zeros(employees, dtype=int)
for i in range(employees):
    total = 0
    for j in range(months):
        total += sales[i][j]
    employee_totals[i] = total

# Calculate total sales for each month
month_totals = np.zeros(months, dtype=int)
for j in range(months):
    total = 0
    for i in range(employees):
        total += sales[i][j]
    month_totals[j] = total

# Calculate total sales of the organization
total_sales = 0
for i in range(employees):
    total_sales += employee_totals[i]

print("\nTotal sales of each employee:")
for i in range(employees):
    print(f"Employee {i+1}: {employee_totals[i]}")

print("\nTotal sales of each month:")
for j in range(months):
    print(f"Month {j+1}: {month_totals[j]}")

print("\nTotal sales of the organization:", total_sales)
```

```python
import numpy as np

def determinant(matrix, n):
    if n == 1:
        return matrix[0][0] 
    
    if n == 2:  
        return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
    
    det = 0
    for col in range(n):
        submatrix = np.zeros((n-1, n-1), dtype=int)
        for i in range(1, n):
            sub_col = 0
            for j in range(n):
                if j == col:
                    continue
                submatrix[i-1][sub_col] = matrix[i][j]
                sub_col += 1
        
        cofactor = ((-1) ** col) * matrix[0][col] * determinant(submatrix, n-1)
        det += cofactor
    
    return det

N = int(input("Enter the size of the square matrix: "))

matrix = np.zeros((N, N), dtype=int)

print("Enter the elements of the matrix row by row:")
for i in range(N):
    for j in range(N):
        matrix[i][j] = int(input())

det = determinant(matrix, N)

print("Determinant of the matrix:", det)
```

```python
CITIES = 20
DAYS = 31

temperatures = []
for i in range(CITIES):
    print(f"Enter temperatures for City {i+1}:")
    city_temps = [int(input()) for _ in range(DAYS)]
    temperatures.append(city_temps)


for i in range(CITIES):
    city_temps = temperatures[i]
    
    
    highest_temp = city_temps[0]
    day_of_highest = 1  

    for d in range(1, DAYS):
        if city_temps[d] > highest_temp:
            highest_temp = city_temps[d]
            day_of_highest = d + 1  

    avg_temp = sum(city_temps) / DAYS 
    
    print(f"City {i+1}: Highest Temp = {highest_temp}, Recorded on Day {day_of_highest}, Average Temp = {avg_temp:.2f}")
```
